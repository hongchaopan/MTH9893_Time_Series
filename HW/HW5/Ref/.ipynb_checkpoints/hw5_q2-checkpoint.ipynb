## Requires pykalman python package to run

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pykalman import KalmanFilter
```

```python
input_data = pd.read_csv('raw_prices_aapl_spx.csv')
```

```python
# Calculate percent changes to get returns
input_data['aapl_returns'] = input_data['AAPL_PX_LAST'].pct_change()
input_data['spx_returns'] = input_data['SPX_PX_LAST'].pct_change()
input_data.dropna(inplace=True)
```

```python
# the function is deprecated but will work for our purposes as no similar alternative exists
# within statsmodels
moving_ols_results = pd.stats.ols.MovingOLS(y=input_data['aapl_returns'],
                      x = input_data['spx_returns'],
                      window_type='rolling',
                      window=63)
```

```python
# Extract out the alpha, beta and residuals from the rolling regression
alpha = moving_ols_results.beta['intercept']
beta = moving_ols_results.beta['x']
residuals = moving_ols_results.resid

# join the results fitted values from the regression to our original frame
input_data['fitted_aapl_returns'] = moving_ols_results.y_fitted
```

```python
# Inputs for state space
X_t = alpha
# y_fitted is the result of beta*m_t + alpha. 
# As a result the subtraction gives us - (beta*m_t + alpha) so we add back on alpha
Y_t = input_data['aapl_returns'] - input_data['fitted_aapl_returns'] + alpha
```

```python
# Make a combined frame of our X_t and Y_t
joined_data_frame = pd.concat((X_t, Y_t), axis=1)
joined_data_frame.columns = ['X_t', 'Y_t']
joined_data_frame.dropna(inplace=True)
```

### Construct and run Kalman filter

```python
# Referred to http://www.thealgoengineer.com/2014/online_linear_regression_kalman_filter/
observation_matrix = np.vstack([joined_data_frame['X_t'],
                                np.ones(joined_data_frame['X_t'].shape)]).T[:, np.newaxis]
```

```python
delta = 1E-5
transition_covariance = delta/ (1 - delta)*np.eye(2)
```

```python
kalman_filter = KalmanFilter(initial_state_mean=np.zeros(2),
                            transition_matrices=np.eye(2),
                            observation_matrices=observation_matrix,
                            observation_covariance=1.0,
                            transition_covariance=transition_covariance
                            )
```

### Run Kalman filter to model the $\alpha_{t}$

```python
state_means, state_covs = kalman_filter.filter(joined_data_frame['X_t'].values)
state_means = pd.DataFrame(state_means)
```

```python
plt.plot(state_means.index, state_means[1],
         'r--', label='Kalman Filter value')
plt.plot(state_means.index, joined_data_frame['X_t'].values,
         'b--',label='Original value')
plt.xlabel('Index of observation')
plt.ylabel(r'$ \alpha$ value')
plt.title(r'Kalman Filter $ \alpha$ and original $ \alpha$')
plt.legend()
plt.show()
```

### Calculate the smoothed values of $\alpha_{t}$ and plot$

```python
smooth_means, smooth_covs = kalman_filter.smooth(joined_data_frame['X_t'].values)
smooth_means = pd.DataFrame(smooth_means)
```

```python
plt.plot(smooth_means.index, smooth_means[1],'r--',label='Smoothed value')
plt.xlabel('Index of observation')
plt.ylabel(r'$ \alpha$ value')
plt.title(r'Smoothed Kalman Filter $ \alpha$')
plt.legend()
plt.show()
```
